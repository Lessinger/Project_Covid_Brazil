# Brazil and COVID-19, analysing the big picture

COVID-19 is an infectious disease caused by newly discovered coronavirus.

Transmitted mainly through droplets from coughs or sneezes of infected people, the severity of symptoms varies greatly from person to person.

The fact is, not much is known about COVID-19. Studies are being carried out around the world, but the results are not yet conclusive and definitive.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

To date, it has been observed that around 80% of confirmed cases are asymptomatic and rapid. Most people who fall into this group recover without any sequelae.

However, 15% of people will have serious infections and require oxygen. The rest of the people, which represent 5%, will be classified as very serious infections and will need assisted ventilation, using mechanical respirators in a hospital environment.

With the aim of raising situational awareness regarding COVID-19 in Brazil, I will carry out an analysis of public data on the disease.

## Data collection

*All the data used on this analysy came from [Our World in Data](https://ourworldindata.org/) and it's public available to anyone intersting in perform similar analisys. Through several diferent databases, 13,231 charts across 120 topics, as informed below the search bar of the website.*

*During the pandemic, some different data was available online. These data don't match each other and some unofficial data also came to attention of people. The date here was collected from the [World Health Organization](https://data.who.int/dashboards/covid19/cases?n=c).*

### Variables dictionary

* `country` - Name of the country
* `date` - Date related to the data

#### Regarding Cases:
* `total_cases` - Total confirmed cases of COVID-19.
* `new_cases` - New confirmed cases of COVID-19.
* `new_cases_smoothed` - New confirmed cases of COVID-19
* `total_cases_per_million` - Total confirmed cases of COVID-19 per 1,000,000 people.
* `new_cases_per_million` - New confirmed cases of COVID-19 per 1,000,000 people.
* `new_cases_smoothed_per_million` - New confirmed cases of COVID-19 per 1,000,000 people.*

<sub> *All `_smoothed` represents 7-day smoothed </sub>

#### Regarding Deaths
* `total_deaths` - Total deaths attributed to COVID-19.
* `new_deaths` - New deaths attributed to COVID-19
* `new_deaths_smoothed` - New deaths attributed to COVID-19*
* `total_deaths_per_million` - Total deaths attributed to COVID-19 per 1,000,000 people.
* `new_deaths_per_million` - New deaths attributed to COVID-19 per 1,000,000 people.
* `new_deaths_smoothed_per_million` - New deaths attributed to COVID-19*

<sub> *All `_smoothed` represents 7-day smoothed </sub>

<sub>Official note from the repository: 'Due to varying protocols and challenges in the attribution of the cause of death, the number of confirmed deaths may not accurately represent the true number of deaths caused by COVID-19.'</sub>

#### Regarding Mortality
* `excess_mortality` - Percentage difference between the reported number of weekly or monthly deaths in 2020–2021 and the projected number of deaths for the same period based on previous years.
* `excess_mortality_cumulative` - Percentage difference between the cumulative number of deaths since 1 January 2020 and the cumulative projected deaths for the same period based on previous years.
* `excess_mortality_cumulative_absolute` - Cumulative difference between the reported number of deaths since 1 January 2020 and the projected number of deaths for the same period based on previous years.
* `excess_mortality_cumulative_per_million` - Cumulative difference between the reported number of deaths since 1 January 2020 and the projected number of deaths for the same period based on previous years, per million people.

#### Regarding Hospital & Intensive Care Unit
* `icu_patients` - Number of COVID-19 patients in intensive care units (ICUs) on a given day
* `icu_patients_per_million` - Number of COVID-19 patients in intensive care units (ICUs) on a given day per 1,000,000 people
* `hosp_patients` - Number of COVID-19 patients in hospital on a given day
* `hosp_patients_per_million` - Number of COVID-19 patients in hospital on a given day per 1,000,000 people
* `weekly_icu_admissions` - Number of COVID-19 patients newly admitted to intensive care units (ICUs) in a given week
* `weekly_icu_admissions_per_million` - Number of COVID-19 patients newly admitted to intensive care units (ICUs) in a given week per 1,000,000 people
* `weekly_hosp_admissions` - Number of COVID-19 patients newly admitted to hospitals in a given week
* `weekly_hosp_admissions_per_million` - Number of COVID-19 patients newly admitted to hospitals in a given week per 1,000,000 people

#### Regarding Policy responses
* `stringency_index` - Government Response Stringency Index*

 <sub>*Composite measure based on 9 response indicators including school closures, workplace closures, and travel bans, rescaled to a value from 0 to 100 (100 = strictest response)</sub>

#### Regarding Reproduction rate
* `reproduction_rate` - Real-time estimate of the effective reproduction rate (R) of COVID-19

#### Regarding Tests & positivity
* `total_tests` - Total tests for COVID-19
* `new_tests` - New tests for COVID-19
* `total_tests_per_thousand` - Total tests for COVID-19 per 1,000 people
* `new_tests_per_thousand` - New tests for COVID-19 per 1,000 people
* `new_tests_smoothed` - New tests for COVID-19**
* `new_tests_smoothed_per_thousand` - New tests for COVID-19**
* `positive_rate` - The share of COVID-19 tests that are positive*
* `tests_per_case` - Tests conducted per new confirmed case of COVID-19*
* `tests_units` - Units used by the location to report its testing data

<sub> *`positive_rate` and `tests_per_case` are the inverse of each other </sub>
<sub> **All `_smoothed` represents 7-day smoothed </sub>

#### Regarding Vaccinations
* `total_vaccinations` - Total number of COVID-19 vaccination doses administered
* `people_vaccinated` - Total number of people who received at least one vaccine dose
* `people_fully_vaccinated` - Total number of people who received all doses prescribed by the initial vaccination protocol
* `total_boosters` - Total number of COVID-19 vaccination booster doses administered
* `new_vaccinations` - New COVID-19 vaccination doses administered (only calculated for consecutive days)
* `new_vaccinations_smoothed` - New COVID-19 vaccination doses administered*
* `total_vaccinations_per_hundred` - Total number of COVID-19 vaccination doses administered per 100 people in the total population
* `people_vaccinated_per_hundred` - Total number of people who received at least one vaccine dose per 100 people in the total population
* `people_fully_vaccinated_per_hundred` - Total number of people who received all doses prescribed by the initial vaccination protocol per 100 people in the total population
* `total_boosters_per_hundred` - Total number of COVID-19 vaccination booster doses administered per 100 people in the total population
* `new_vaccinations_smoothed_per_million` - New COVID-19 vaccination doses administered per 1,000,000 people in the total population*
* `new_people_vaccinated_smoothed` - 	Daily number of people receiving their first vaccine dose*
* `new_people_vaccinated_smoothed_per_hundred` - Daily number of people receiving their first vaccine dose per 100 people in the total population*

<sub> *All `_smoothed` represents 7-day smoothed </sub>

#### Others

### Import process of the COVID-19 data
*The dataset have around 100 megabytes of raw data, and it's already available to be used by the [Github of Our World in Data](https://github.com/owid/covid-19-data/tree/master/public/data), the extension used on this analysis is as `csv` format.*




```
# import the needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import the dataset to create the dataframe
df = pd.read_csv('https://catalog.ourworldindata.org/garden/covid/latest/compact/compact.csv')
```

* *As verified above, the column `life_expectancy` is all blank, `weekly_icu_admissions_per_million`, `weekly_icu_admissions`, `excess_mortality_cumulative_per_million` and `excess_mortality` have around 97% of the data blank and on the other point of the table we have `country` and `date` with 0% of the data blank followed by `total_cases_per_million`, `total_deaths_per_million` and `total_cases` that have around 2% of blank data.*

* *It's important to highlight here that as the beguinning of the data collection was 2020/01/01, some countries had not report any cases during this started period, explaining why some of the `total_cases` are blank, even that some of the similiar cases are filled with 0.*

* *Just to confirm the theory, I'll perform an `.loc` funtion, that returns all the lines that have `0` on the `total_cases` as the system understands it as `False`, and with this slice of dataframe I'll run another test after that.*

```
(df.loc[df["total_cases"] == False])
```

```
(df.loc[df.total_cases == False]).isnull().sum()
```

* *The theory was confirmed above,  as the slicing process created wasn't able to find any NoN-Null information on `total_cases`*

* *Changing the datatype of `date` series for the format `datetime`, during the `df.info()` command it showed that `date` was `Object` and this can cause some issues during our plotting.*

```
df.date = pd.to_datetime(df.date)
```

* *Using again the `loc` funtion, now related to an specific date that it might be interesting to find, as example I'll check for `2024/01/16`, exactly one year ago. For this date, I'll locate by sorting the numbers of `total_cases` and `total_deaths`.*


```
df.loc[df.date == '2024-01-16'].sort_values(by="total_cases", ascending=False)
```

```
df.loc[df.date == '2024-01-16'].sort_values(by="total_deaths", ascending=False)
```

* Now, the proposal is to plot a graph considering the top 5 countries with most number os deaths, in relation to the previously date that was sliced from the dataframe. Note that the scale is, for exemple, **7e6** which means **7 000 000** deaths



```
fig, ax = plt.subplots()
top5_total_deaths = df.loc[df.date == '2024-01-16', ['country', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]
ax.bar('country', 'total_deaths', data = top5_total_deaths)
ax.set_title('Top 5 countries with most deaths')
ax.set_xlabel('Country')
ax.set_ylabel('Total deaths')
ax.set_xticklabels(top5_total_deaths['country'], rotation=90)
plt.show()
```

* So, now I'll plot a graph considering the total number of deaths around the world according to the days elapsed. Note that, again, the scale is **7e6** which means **7 000 000** deaths


```
fig, ax = plt.subplots()
df.loc[df.country == "World", "total_deaths"].reset_index(drop=True).plot()
ax.set_title('Total deaths in the World')
ax.set_xlabel('Elapsed days')
ax.set_ylabel('Total deaths')
plt.show()
```






* Above, I have plotted an scatter graph showing `gdp_per_capita` and `total_deaths` for the date of **2024-01-16**.
   


```
fig, ax = plt.subplots()
gdp_total_deaths = df.loc[df.date == '2024-01-16', ['gdp_per_capita', 'total_deaths']]
ax.scatter('gdp_per_capita', 'total_deaths', data = gdp_total_deaths)
ax.set_title('GDP per capita vs. total deaths')
ax.set_xlabel('GDP per capita')
ax.set_ylabel('Total deaths')
plt.show()
```

### Brazil's exploratory analysis

* In order to work only with the data from Brazil, a new dataframe will be generated from the original with the funtion `.copy`, together with the `.loc` funtion to perform the desired slice on the dataframe.

* From the `df_brazil`, I'll perform the analysis of the first `total_case`, after, the first `total_death` and the period between both.


```
df_brazil = df.loc[df.country == "Brazil"].copy()
```

```
df_brazil.head()
```

```
df_brazil.loc[df.total_cases == True].sort_values(by="date").head(1)
```

```
df_brazil.loc[df_brazil.total_deaths == True].sort_values(by="date").head(1)
```

* For the last two parts of code, the first one shows the fist `date` that Brazil notify a `total_cases`, **2020-02-27**, and the second line inform the fist `date` that Brazil notify the first `total_deaths`, **2020-03-19**.
* According to the dataset I'm working on, **20 days** pass from Brazil's first confirm case to the first confirm death.

---

* Now, an analisys searching for the top five days with higher number of `new_cases`.
* As the code shows, the top five days happened between the end of January 2022 and the beginning of February 2022, related to the *Second Wave*, below the table for `new cases` I'll plot and graph with the full series of `new_cases` related to `date`. It'll be clear to see the peak of `new_cases` on the starting period of **2022**.

```
df_brazil.sort_values(by="new_cases", ascending=False).head(5)
```

```
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10,4))

y1 = df_brazil.new_cases
x1 = df_brazil.date

# gráfico 1
ax1.plot(x1, y1)
ax1.set_title("Brazil new cases")
ax1.set_xlabel("Year")
ax1.set_ylabel("New Cases")
```

* Before final considerations, two more plots that demonstrate the evolution of `total_cases` and `total_deaths` during the years in `Brazil`. After, the same kind of graph but now with the `World` data, it's possible to compare how similar both charts are mainly when we compare `total_deaths`.

```
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

y1 = df_brazil.total_cases
x1 = df_brazil.date
y = df_brazil.total_deaths

# gráfico 1
ax1.plot(x1, y1)
ax1.set_title("Brazil total cases")
ax1.set_xlabel("Year")
ax1.set_ylabel("Total Cases")

# gráfico 2
ax2.plot(x1, y)
ax2.set_title("Brazil total deaths")
ax2.set_xlabel("Year")
ax2.set_ylabel("Total Deaths")

fig.tight_layout();
```

```
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

y1 = df.loc[df.country == "World"].total_cases
x1 = df.loc[df.country == "World"].date
y = df.loc[df.country == "World"].total_deaths

# 1st graph
ax1.plot(x1, y1)
ax1.set_title("World total cases")
ax1.set_xlabel("Year")
ax1.set_ylabel("Total Cases")

# 2nd graph
ax2.plot(x1, y)
ax2.set_title("World total deaths")
ax2.set_xlabel("Year")
ax2.set_ylabel("Total Deaths")

fig.tight_layout();
```
